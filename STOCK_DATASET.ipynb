```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from scipy import stats
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
stock=pd.read_csv("Stock.csv")
print(stock)
```

```python
stock['Volume'].mean()
```

```python
stock=stock.dropna()
```

```python
missing_values = stock.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
stock['Date'] = pd.to_datetime(stock['Date'])
```

```python
stock = stock.sort_values(by='Date')
```

```python
stock = stock.drop_duplicates()
```

```python
stock= stock.reset_index(drop=True)
```

```python
stock['Date'].unique().shape
```

```python
stock.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(stock['Volume'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(stock['Open'], stock['Close'], color='orange')
plt.title('Scatterplot of Open vs Close')
plt.xlabel('Open')
plt.ylabel('Close')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(stock.index, stock['Open'], color='blue')
plt.title('Open Prices Over Time[Line plot]')
plt.xlabel('Date')
plt.ylabel('Open Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sb.boxplot(y=stock['High'], color='green')
plt.title('Box Plot of High Prices')
plt.ylabel('High Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sb.boxplot(x=stock['High'], color='green')
plt.title('Box Plot of High Prices')
plt.ylabel('High Price')
plt.show()
```

```python
sb.pairplot(stock[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']])
plt.suptitle('Pair Plot of Open, High, Low, Close, Adj Close, and Volume', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sb.heatmap(stock[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Between Columns')
plt.show()
```
